```python
import pandas as pd

def read_data(key):
    """
    Read data from a CSV file based on the SELECT query.

    Parameters:
    - key (str): The SELECT query.

    Returns:
    - None: Prints the selected data or error messages.
    """
    try:
        columns_to_fetch = key.split('from')[0].split('select')[1].strip()
        if "where" in key:
            dataframe_name = key.split('where')[0].split('from')[1].strip()
            df = pd.read_csv(f"{dataframe_name}.csv")
            column_name = key.split('where')[1].split('=')[0].strip()
            value = key.split('where')[1].split('=')[1].strip()
            value = "".join(c for c in value if c.isalpha())
            if columns_to_fetch == "*":
                print(df[df[column_name] == value])
        else:
            dataframe_name = key.split('from')[1].strip()
            dataframe_name = "".join(c for c in dataframe_name if c.isalnum())
            df = pd.read_csv(f"{dataframe_name}.csv")
            if columns_to_fetch == "*":
                print(df[:11])
    except KeyError as e:
        print(f"Error: {e} column(s) not found.")
    except FileNotFoundError as e:
        print(f"Error: File {e.filename} not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

def extract_column(key):
    """
    Extract column names from the INSERT query.

    Parameters:
    - key (str): The INSERT query.

    Returns:
    - List: Extracted column names.
    """
    try:
        columns_start_index = key.find('(') + 1
        columns_end_index = key.find(')')
        column_str = key[columns_start_index:columns_end_index].strip()
        column_names = [col.strip() for col in column_str.split(',')]
        return column_names
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

def extract_values(insert_statement):
    """
    Extract values from the INSERT query.

    Parameters:
    - insert_statement (str): The INSERT query.

    Returns:
    - List: Extracted values.
    """
    try:
        values = insert_statement.split("values(")[1]
        values = values.split(")")[0].strip().split(",")
        cleaned_values = ["".join(c for c in val if c.isalnum()) for val in values]
        return cleaned_values
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

def insert_data(key):
    """
    Insert data into a CSV file based on the INSERT query.

    Parameters:
    - key (str): The INSERT query.

    Returns:
    - None: Prints success message or error messages.
    """
    try:
        column_names = extract_column(key)
        values = extract_values(key)
        dataframe_name = key.split("into")[1]
        dataframe_name = dataframe_name.split("(")[0].strip()
        df = pd.read_csv(f"{dataframe_name}.csv")
        df = df.append({column_names[i]: values[i] for i in range(len(column_names))}, ignore_index=True)
        df.to_csv(f"{dataframe_name}.csv", index=False)
    except FileNotFoundError as e:
        print(f"Error: File {e.filename} not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

def delete_data(key):
    """
    Delete rows from a CSV file based on the DELETE query.

    Parameters:
    - key (str): The DELETE query.

    Returns:
    - None: Prints success message or error messages.
    """
    try:
        if "where" in key:

            dataframe_name = key.split('from')[1].split('where')[0].strip()
            dataframe_name = "".join(c for c in dataframe_name if c.isalnum())
            df = pd.read_csv(f"{dataframe_name}.csv")
            print(dataframe_name)

            where_clause = key.split('where')[1].strip()
            column_name, value = where_clause.split('=')
            column_name = column_name.strip()
            value = value.strip().strip("'")
            value="".join(c for c in value if c.isalnum())
            rows_to_delete = df[df[column_name] == value]
            print(rows_to_delete)

            if not rows_to_delete.empty:
                df = df.drop(rows_to_delete.index)
                df.to_csv(f"{dataframe_name}.csv", index=False)
                print(f"Deleted rows where {column_name} is {value}")
            else:
                print(f"No rows matched the condition: {column_name} = {value}")
        else:
            print("DELETE statement must include a WHERE clause.")
    except FileNotFoundError as e:
        print(f"Error: File {e.filename} not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

```

```python
def sql_parser():
  while True:
    print()
    key = input().lower()
    if "select" in key:
        read_data(key)
    elif "insert" in key:
        insert_data(key)
    elif "delete" in key:
      delete_data(key)
    if key=="exit":
      break


sql_parser()
```
